```python
import pandas as pd
import plotly.express as px

file_path = 'cardio_train.csv'
df = pd.read_csv(file_path, sep = ';')

# .size() counts the rows in each group and reset_index() turns it into a new DataFrame
df_agg = df.groupby(['cholesterol', 'cardio']).size().reset_index(name='# of patients')

# Map the numbers to meaningful strings for a cleaner chart
df_agg['Cholesterol Level'] = df_agg['cholesterol'].map({
    1: '1: Normal',
    2: '2: Above Normal',
    3: '3: Well Above Normal'
})

df_agg['Presence of CVD'] = df_agg['cardio'].map({
    0: '0: No CVD',
    1: '1: Has CVD'
})

# Display the new, aggregated DataFrame
print("This is the data being plotted:")
print(df_agg)

# Now use the aggregated DataFrame 'df_agg'
fig = px.bar(
    df_agg,
    x="Cholesterol Level",
    y="# of patients",
    color="Presence of CVD",
    title="Patient Count by Cholesterol Level and CVD Status"
)

fig.show()
```
