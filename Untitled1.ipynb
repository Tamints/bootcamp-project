```python
import pandas as pd
import numpy as np

#loading data set

df = pd.read_csv("cardio_train - cardio_train.csv", sep = ",")

```

```python
# Coverting days to years
df["age_years"] = (df["age"]/ 365). round().astype(int)
```

```python
# finding BMI - Weight or heigh alon not predictive, but heigh might be predictive of blood pressure
df["BMI"] = df["weight"] / (df["height"] * df["height"] / 10000)
```

```python
#data cleaning BMI, HEIGHT, BP ETC
# Remove unrealistic blood pressure values
df.drop(df[(df['ap_hi'] > 250) |
            (df['ap_hi'] < 50)].index, inplace=True)

df.drop(df[(df['ap_lo'] > 140) |
            (df['ap_lo'] < 30)].index, inplace=True)

df.drop(df[(df['ap_hi'] < df['ap_lo']) |
            (df['ap_lo'] > df['ap_hi'].quantile(0.025))].index, inplace=True)
```

```python
#Mean arterial pressure
df["MAP"] = (df["ap_hi"]) /3 + 2 * (df["ap_lo"]) / 3
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate the correlation matrix on your cleaned and engineered data
# (Make sure to include 'age_years', 'bmi', and 'cardio')
corr_matrix = df[['age_years', 'BMI', 'cholesterol', 'gluc', 'smoke', 'alco', 'active', 'MAP', 'cardio', ]].corr()

# Plot the heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, cmap='viridis', fmt='.2f')
plt.title('Correlation Heatmap of CVD Factors')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a figure and axes
fig, ax = plt.subplots(figsize=(12, 8))

# Create the violin plot
# 'x' is your categorical feature (cholesterol)
# 'y' is your continuous feature (systolic blood pressure)
# 'hue' splits the violins by your target (cardio)
# 'split=True' creates the "half-and-half" violin for a direct comparison
sns.violinplot(data=df, x='cholesterol', y='MAP', hue='cardio',
               split=True, inner='quartile', palette='Set1', ax=ax)

plt.title('Systolic Blood Pressure Distribution by Cholesterol and CVD Status')
plt.xlabel('Cholesterol Level (1: norm, 2: above, 3: well above)')
plt.ylabel('Mean Arterial Blood Pressure (MAP)')
plt.legend(title='Cardio', loc='upper left', labels=['No CVD', 'CVD'])
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# --- Load and Prepare Data (Assuming 'df' is loaded and cleaned) ---
# This step assumes 'df' is your DataFrame with outliers removed
# and 'age_years' and 'bmi' are engineered.

# 1. Create the 'lifestyle' interaction feature
def get_lifestyle_group(row):
    if row['smoke'] == 0 and row['alco'] == 0:
        return 'No Smoke, No Alcohol'
    if row['smoke'] == 1 and row['alco'] == 0:
        return 'Smoker, No Alcohol'
    if row['smoke'] == 0 and row['alco'] == 1:
        return 'Non-Smoker, Alcohol'
    if row['smoke'] == 1 and row['alco'] == 1:
        return 'Smoker, Alcohol'

df['lifestyle'] = df.apply(get_lifestyle_group, axis=1)

# 2. Create the plot
# We use sns.catplot (kind='bar') to show the mean of 'cardio'
# for each group. The 'ci=None' is now 'errorbar=None'.
g = sns.catplot(
    data=df,
    x='lifestyle',
    y='cardio',
    kind='bar',
    palette='Set1',
    order=['No Smoke, No Alcohol', 'Non-Smoker, Alcohol', 'Smoker, No Alcohol', 'Smoker, Alcohol']
)

g.fig.suptitle('CVD Prevalence by Smoking and Alcohol Interaction', y=1.03)
g.set(xlabel='Lifestyle Group', ylabel='Rate of Cardiovascular Disease (Mean)')
plt.xticks(rotation=15)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# --- Load and Prepare Data (Assuming 'df' is loaded) ---

# 1. Create a pivot table of the MEAN cardio risk for each combination
# We use.pivot_table() to calculate the mean(cardio) for each cell
chol_gluc_pivot = df.pivot_table(
    values='cardio',
    index='cholesterol',
    columns='gluc',
    aggfunc='mean'
)

# 2. Create the heatmap from this pivot table
plt.figure(figsize=(10, 7))
sns.heatmap(
    chol_gluc_pivot,
    annot=True,       # Show the percentages in each cell
    fmt='.1%',        # Format as percentage
    cmap='OrRd',      # Use an Orange-Red color map
    linewidths=.5
)

plt.title('CVD Rate by Cholesterol and Glucose Levels', fontsize=16)
plt.xlabel('Glucose Level (1:Normal, 2:Above, 3:Well Above)', fontsize=12)
plt.ylabel('Cholesterol Level (1:Normal, 2:Above, 3:Well Above)', fontsize=12)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# --- Load and Prepare Data (Assuming 'df' is loaded and 'age_years', 'MAP' exist) ---

# 1. Create the faceted scatter plot using sns.relplot
# 'x' and 'y' are continuous variables
# 'hue' is our target (cardio)
# 'col' creates the separate charts for each cholesterol level
g = sns.relplot(
    data=df,
    x='age_years',
    y='MAP',
    hue='cardio',
    col='cholesterol',    # Creates 3 columns: 1, 2, and 3
    palette='coolwarm',
    alpha=0.6,
    height=5,
    aspect=0.8
)

# 2. Set titles and labels
g.fig.suptitle('MAP vs. Age by Cholesterol Level and CVD Status', y=1.05)
g.set_axis_labels('Age (Years)', 'Mean Arterial Pressure (MAP)')
g.legend.set_title('CVD Status')

plt.show()
```

```python
import plotly.express as px
import pandas as pd

# --- Assuming 'df' is loaded and cleaned from the code above ---

# --- 1. Prepare Categorical Data ---
# We need a column to count the size of each group
df['count'] = 1

# Map the numeric categories to readable strings for the chart
df['cardio_status'] = df['cardio'].map({0: 'No CVD', 1: 'Has CVD'})
df['cholesterol_level'] = df['cholesterol'].map({1: 'Normal Chol.', 2: 'Above Normal Chol.', 3: 'High Chol.'})
df['glucose_level'] = df['gluc'].map({1: 'Normal Gluc.', 2: 'Above Normal Gluc.', 3: 'High Gluc.'})

# --- 2. Create the Sunburst Chart ---
# The 'path' creates the hierarchy:
# (Outer Ring) -> (Middle Ring) -> (Inner Ring)
fig_sunburst = px.sunburst(
    df,
    path=['cardio_status', 'cholesterol_level', 'glucose_level'],
    values='count',
    title="Hierarchical Breakdown of Risk Factors (CVD > Cholesterol > Glucose)",

    # We color the chart based on the cholesterol level
    color='cholesterol_level',
    color_discrete_map={
        'Normal Chol.': 'green',
        'Above Normal Chol.': 'orange',
        'High Chol.': 'red'
    }
)

fig_sunburst.show()
```

```python
import plotly.express as px
import pandas as pd
import numpy as np

# --- 4. Map Categories to Strings for Clear Labels ---
# This makes the plot's legend and axes much easier to read
df['cardio_status'] = df['cardio'].map({0: 'No CVD', 1: 'Has CVD'})
df['cholesterol_level'] = df['cholesterol'].map({
    1: '1: Normal',
    2: '2: Above Normal',
    3: '3: Well Above Normal'
})

# --- 5. Create the 3D Interactive Plot ---
# We take a random sample of 5000 points. Plotting all 70,000 in 3D can be slow.
df_sample = df.sample(n=5000, random_state=1)

fig = px.scatter_3d(
    df_sample,
    x='age_years',
    y='MAP',
    z='cholesterol_level',  # Use the new string labels for the Z-axis
    color='cardio_status',  # Color points by CVD status

    title="3D Interaction: Age vs. MAP vs. Cholesterol",

    # Set clear labels for the axes
    labels={
        'age_years': 'Age (Years)',
        'MAP': 'Mean Arterial Pressure (MAP)',
        'cholesterol_level': 'Cholesterol Level'
    },

    # Make the colors intuitive
    color_discrete_map={
        'No CVD': 'blue',
        'Has CVD': 'red'
    },

    # This helps sort the Z-axis in the correct 1, 2, 3 order
    category_orders={"cholesterol_level": ['1: Normal', '2: Above Normal', '3: Well Above Normal']}
)

# Update the markers to be smaller and slightly transparent
# to better see clusters
fig.update_traces(marker=dict(size=3, opacity=0.7))

# Show the interactive plot
fig.show()
```

```python
import plotly.express as px
import pandas as pd
import numpy as np


# --- 3. Aggregate Data for Weak Predictors ---
# "Melt" the subjective features into one column to plot them together
df_melt = df.melt(id_vars=['cardio'], value_vars=['smoke', 'alco', 'active'])

# Calculate the mean CVD rate for each group
df_agg = df_melt.groupby(['variable', 'value'])['cardio'].mean().reset_index()

# Create clearer labels for the chart
df_agg['value_label'] = df_agg['value'].map({0: 'No', 1: 'Yes'})
df_agg['Predictor'] = df_agg['variable'] + ' = ' + df_agg['value_label']

# --- 4. Create the Polar Bar Chart ---
fig1 = px.bar_polar(
    df_agg,
    r='cardio',              # Radius = CVD Rate
    theta='Predictor',       # Angle = Each category
    color='cardio',          # Color by the CVD Rate
    template='plotly_dark',
    title='CVD Rate for "Weak" Subjective Factors',
    color_continuous_scale=px.colors.sequential.Plasma_r,
    labels={'cardio': 'CVD Rate'}
)

fig1.update_layout(
    polar=dict(
        radialaxis=dict(range=[0, 0.55])  # Fix the axis scale
    )
)

fig1.show()
```

```python
import plotly.express as px
import pandas as pd
import numpy as np

# --- 3. Aggregate Data for Strong Predictor ---
# Bin ages into 5-year groups
df['age_bin'] = pd.cut(df['age_years'],
                       bins=range(30, 70, 5),
                       right=False,
                       labels=[f'{i}-{i+4}' for i in range(30, 65, 5)])

# Calculate the mean CVD rate for each age bin
df_agg_age = df.groupby('age_bin', observed=True)['cardio'].mean().reset_index()

# --- 4. Create the Polar Bar Chart ---
fig2 = px.bar_polar(
    df_agg_age,
    r='cardio',
    theta='age_bin',
    color='cardio',
    template='plotly_dark',
    title='CVD Rate by Age Group (A "Strong" Predictor)',
    color_continuous_scale=px.colors.sequential.Plasma_r,
    labels={'cardio': 'CVD Rate', 'age_bin': 'Age Group'}
)

fig2.show()
```
